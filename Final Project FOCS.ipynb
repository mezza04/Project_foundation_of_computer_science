```python
import pandas as pd
import re
import json
import zstandard as zstd
import io
import numpy as np
```

```python
countries = ["CA", "DE", "FR", "GB", "IN", "JP", "KR", "MX", "RU", "US"]
dfs = {}
for element in countries:
    path = fr"C:\Users\chiar\OneDrive\Desktop\foundation of computer science\trendingYT\{element}videos.csv.zst"
    print("Reading:", path)

    with open(path, "rb") as f: # "rb" opens the file in "read binary" mode, it is necessary because the files ar compressed
        dctx = zstd.ZstdDecompressor()
        decompressed = dctx.decompress(f.read()) # decompression using zstd library

        text = decompressed.decode("utf-8", errors="replace") # decoding with utf-8 that is suitable for all lenguages and special characters

        dfs[element] = pd.read_csv(io.StringIO(text)) # dictionary with all datasets

    dfs[element]["country"] = element # adding the "country" column
```

#### 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
df_all_e[["likes", "dislikes"]].eq(0).sum() # we have many 0 in like and dislike that will cause problems
```

```python
df_all_e["like_ratio"] = df_all_e["likes"] / df_all_e["dislikes"] 
pd.Series({
    "inf": np.isinf(df_all_e["like_ratio"]).sum(), # inf happens when dislike = 0 and like is not 0
    "nan": df_all_e["like_ratio"].isna().sum() # nan happens when like and dislike ara 0
})
```

```python
df_all_norm = df_all_e.copy()
df_all_norm["like_ratio"] = df_all_norm["likes"] / (df_all_norm["likes"] + df_all_norm["dislikes"]) # to have no inf and nan we normalize the ratio
df_all_norm["like_ratio"].value_counts()
```

#### 8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.

```python
diz_tag = {}

for riga in df_all_norm["tags"]:
    tags = riga.split("|") # split the string in the different tags
    
    for tag in tags:
        nome_tag = tag.strip().strip('"')
        
        if nome_tag == "" or nome_tag == "[none]": # we skipp [none] as it is not a real tag
            continue
        
        if nome_tag in diz_tag:
            diz_tag[nome_tag] += 1
        else:
            diz_tag[nome_tag] = 1

diz_tag
```

```python
top_10 = sorted(diz_tag.items(), key=lambda x: x[1], reverse=True)[:10]
top_10
```
